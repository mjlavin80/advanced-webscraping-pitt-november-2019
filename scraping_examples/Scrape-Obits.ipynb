```python
import requests, time, os
from calendar import monthrange
from bs4 import BeautifulSoup
```

```python
try:
    os.stat("obits")
except:
    os.mkdir("obits")
try: 
    os.stat("txt")
except:
    os.mkdir("txt")
```

```python
BASE_URL = "https://archive.nytimes.com/www.nytimes.com/learning/general/onthisday/bday/"
TAIL = ".html"
```

```python
for month in range(1,13):
    days = monthrange(2016, month)
    
    for d in range(1, days[1]+1):
        if len(str(month)) == 1:
            month = "0"+str(month)
        if len(str(d)) == 1:
            d = "0"+str(d)
            
        path = "".join(["obits/",str(month), str(d),".html"])
        if not os.path.isfile(path):
            URL = "".join([BASE_URL, str(month), str(d), TAIL])
            data = requests.get(URL)
            with open(path, "a") as f:
                f.write(data.text)
            time.sleep(.5)
        else:
            pass
```

```python
all_obits = []
all_headlines = []
all_dates = []

for month in range(1,13):
    days = monthrange(2016, month)
    for d in range(1, days[1]+1):
        if len(str(month)) == 1:
            month = "0"+str(month)
        if len(str(d)) == 1:
            d = "0"+str(d)
        filename = "".join(["obits/", str(month), str(d), ".html"])
        
        with open(filename) as f:
            html = f.read()
            
        #convert to beautifulsoup object
        soup = BeautifulSoup(html, "html.parser")
        all_ps = []
        
        headline = soup.find_all("h2")
        
        # get headline
        try:
            all_headlines.append(" ".join(headline[0].text.replace("\n", "").split()))
        except:
            all_headlines.append(" ")
        
        #try to find date
        date = soup.find_all("h5")         
        try:
            all_dates.append(date[0].text)
        except:
            all_dates.append(" ")  
        
        #look for p tags in td 
        for i in soup.find_all("td"):
            ps = i.find_all("p")      
            for z in ps:
                try:
                    if z.text not in all_ps:
                        all_ps.append(z.text)
                except:
                    pass   
        obit_text = " ".join(all_ps)
        
        all_obits.append(obit_text)
```

```python
filenames = []
for month in range(1,13):
    days = monthrange(2016, month)
    for d in range(1, days[1]+1):
        if len(str(month)) == 1:
            month = "0"+str(month)
        if len(str(d)) == 1:
            d = "0"+str(d)
        filename = "".join(["txt/", str(month), str(d), ".txt"])
        filenames.append(filename)
```

```python
len(filenames) == len(all_obits)
```

```python
for n, o in enumerate(all_obits):
    o_no_white = " ".join(o.replace("\n", "").replace("&apos", "\'").replace("&apos;", "\'").split())
    with open(filenames[n], "a") as f:
        f.write(o_no_white)
    f.close()
```

```python
import pandas as pd
df_meta = pd.DataFrame()
df_meta['headline'] = all_headlines
df_meta['date'] = all_dates
df_meta['filename'] = filenames
```
